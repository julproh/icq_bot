```python
import numpy as np
import pandas as pd
import xlsxwriter
```

```python
Data = pd.read_excel('jul.xlsx')
Data
```

```python
result = []
col = list(Data.columns)
for j in range(60):
    k = np.random.randint(0, 5, 10)
    tasks = []
    for i in range(10):
        tasks.append(col[i] + ' ' + Data.iloc[k[i], i])
    result.append(tasks)
```

```python
new_names = []
for i in range(1, 11):
    new_names.append('Задача №{}'.format(i))
new_names
```

```python
New = pd.DataFrame(result)
New.columns = new_names
New['Номер варианта контрольной'] = range(1, 61)
New
```

```python
writer = pd.ExcelWriter('Jul_outcomes.xlsx', engine='xlsxwriter')
New.to_excel(writer, sheet_name='Sheet1')
writer.save()
```
